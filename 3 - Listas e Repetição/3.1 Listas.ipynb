Quando trabalhamos com conjunto de dados, precisamos armazenar-los na memória para poder recuperar e manipular-los depois, ou seja, poderiamos armazenar cada dado em uma variável. Abaixo temos um exemplo de como podemos armazenar dados na memória:

### 3.1 Listas
Uma lista pode conter varios ou o mesmo tipo de dados. Uma lista como [4,5,6] tem o mesmo tipo de dado (Inteiros), enquanto que a lista ['Facebook', 0.0, 'USD', 2974676, 3.5] contém dados do tipo:

- 2 strings ("Facebook","USD")
- 2 floats (0.0,3.5)
- 1 inteiro (2974676)

Para encontrar o tamanhao de uma lista podemos usar o comando **len()**:

[In]
```python
row_1=['Facebook', 0.0, 'USD', 2974676, 3.5]
print(len(row_1))
list_1=[1,5,9]
print(len(list_1))

list_2=[]
print(len(list_2))
```
[Out]

5

3

0

Para listas pequenas basta contar os pontos de dados, porém o comando **len()** é bastante eficiente quando a lista é muito grande.

Cada elemento da lista (Data points) é associado a um número específico, chamado de **index number**(Número de Indice), que indica sua posição na lista. Os indices sempre começam no 0, o segundo elemento com indice 1, assim por diante.
```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
              0       1     2       3      4
                            #Indices
```
Uma forma de saber a posição de um elemento da lista, identificar sua posição na lista e subtrair 1. Por exemplo, a sting "USD" é o terceiro elemento na lista, para achar  seu indice basta fazer a subtração, 3-1=2.

Saber o indice de um número da lista ajuda a isolar este elemento. Assim na lista **row_1**, por exemplo, o primeiro elemento ("Facebook") com o indice 0 pode ser chamado pelo código **row_1[0]**.
[In]
```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
row_1[0]
```
[Out]

'Facebook'

A sintaxe para isolar o elemento individual da lista tem como modelo **Nome_lista[Numero_Indice]**.

Abaixo tem uma forma de recuperar cada elemento da lista **row_1**.
[In]
```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
print(row_1[0])
print(row_1[1])
print(row_1[2])
print(row_1[3])
print(row_1[4])
```
[Out]

'Facebook'

0.0

'USD'

2974676
