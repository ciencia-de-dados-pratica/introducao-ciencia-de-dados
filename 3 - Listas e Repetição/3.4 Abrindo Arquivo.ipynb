### 3.4 Abrindo Arquivo

Aprendemos a criar um data_set(Conjunto de dados) apartir de algumas listas de Row's (linhas), entretanto para trabalharmos com mais dados precisaríamos escrever cada dado em uma row diferente e isso tomaria muito tempo e esforço. Por isso, vamos aprender a usar data_set's ja prontos.

Primeiro precisamos encontrar um data_set que queremos trabalhar, no site abaxo pode-se encontrar uma vasta quantidade de data_set's.

https://www.kaggle.com/datasets

Vamos trabalhar com o arquivo **AppleStore.csv** e para importar esses dados faz-se necessário o uso de alguns comandos. A extensão csv é referente a arquivos de texto criados apartir de uma planilha, como o Excel.

- Em arquivos existem quatro operações funcamentais, são elas: Abertura, Fechamento, Leitura e Escrita;
- O primeiro passo que temos que fazer é **Abrir** um arquivo, usando o comando **open** e salvando esses dados em uma variável;
```python
open_file = open("AppleStore.csv")
```
- É necessário poder  **Ler** os dados do arquivo .csv, assim usamosfunção **reader** da biblioteca **csv**, para poder ler os arquivos de forma que os dados sejam, organizadamente, importados.
```python
from csv import reader
read_file = reader(open_file)
```
- Agora faz-se necessário transformar os dados lidos em uma lista de lista, ou seja o nosso data_set, usando o comando **list**.

```python
apps_data = list(read_file)
```
- Por ultimo, é do nosso interesse que o comando **len** seja aprendido. Este comando retorna a quantidade de elementos que existem na lista.

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
len(row_1)

```
[Out] 5

````python
open_file = open("AppleStore.csv")
from csv import reader
read_file = reader(open_file)
apps_data = list(read_file)
print(len(apps_data))
apps_data[:30]
````
